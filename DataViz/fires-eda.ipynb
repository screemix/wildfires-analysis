```python
import pandas as pd
```

```python
data_path = 'data/thermopoints.csv'
df = pd.read_csv(data_path, sep=';')
df
```

I needed only a year of each fire to aggregate annual data

```python
df['year'] = df['dt'].apply(lambda x: x.split("-")[0])
```

```python
df['year'] = df['dt'].apply(lambda x: int(x[:4]))
df
```

```python
df['dt'] = df['dt'].apply(pd.to_datetime)
```

Checking unique categories of fires that persist in the dataaset

```python
df.type_name.unique()
```

```python
df.dtypes
```

Making a table aggregated by category through the whole time

```python
df_categories = df[['type_name', 'dt']].groupby(by=['type_name']).count()
df_categories
```

```python
df_categories.to_csv('overall_categories.csv')
```

Making a table aggregated by both category and year columns for the main part of composition

```python
df = df[['year', 'type_name', 'dt']].groupby(by=['year', 'type_name']).count()
```

```python
df
```

```python
df.to_csv('preprocessed_dataset.csv')
```
